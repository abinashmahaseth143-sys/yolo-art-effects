```python
# notebooks/1_dataset_collection.ipynb
import os
import requests
from PIL import Image
import json

# Create categories based on teacher's examples
categories = {
    'classical_paintings': [
        'https://upload.wikimedia.org/wikipedia/commons/thumb/e/ea/Van_Gogh_-_Starry_Night_-_Google_Art_Project.jpg/1280px-Van_Gogh_-_Starry_Night_-_Google_Art_Project.jpg',
        'https://upload.wikimedia.org/wikipedia/commons/thumb/0/0f/1665_Girl_with_a_Pearl_Earring.jpg/800px-1665_Girl_with_a_Pearl_Earring.jpg'
    ],
    'modern_art': [
        'urls_of_modern_art',
    ],
    'personal_photos': []  # You'll add your own
}

# Download function
def download_images(category, urls, save_dir='data/images'):
    os.makedirs(f'{save_dir}/{category}', exist_ok=True)
    
    for i, url in enumerate(urls):
        try:
            response = requests.get(url, timeout=10)
            if response.status_code == 200:
                with open(f'{save_dir}/{category}/{category}_{i:03d}.jpg', 'wb') as f:
                    f.write(response.content)
                print(f"Downloaded: {category}_{i:03d}.jpg")
        except:
            continue

# Download samples
for category, urls in categories.items():
    if urls:  # Skip empty categories
        download_images(category, urls[:5])  # Limit to 5 per category

print("Dataset collection complete!")
```
